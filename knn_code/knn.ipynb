```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.neighbors import BallTree

rng = np.random.default_rng(42)
X = rng.integers(low=0, high=10, size=(10, 2))

ball_tree = BallTree(X, leaf_size=2)
_, indices, circles, _ = ball_tree.get_arrays()
```

```python
from matplotlib.patches import Circle

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot()

ax.plot(X[:, 0], X[:, 1], 'xb')
ax.grid(True)
plt.xlabel('x')
plt.ylabel('y')
for x, y in X:
    ax.text(x, y, f'({x}, {y})', fontsize=12,
             verticalalignment='top', horizontalalignment='left')


for idx, circ in zip(indices, circles):
    center = (X[idx, 0], X[idx, 1])
    radius = circ[-1]
    ax.add_patch(Circle(center, radius=radius, ec='black'))

plt.show()
```
